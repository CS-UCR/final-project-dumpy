```python
import math
import pandas as pd

# from aequitas.group import Group
# from aequitas.bias import Bias
# from aequitas.fairness import Fairness
# import aequitas.plot as ap

import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import hvplot.pandas
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("dataset.csv")
df
```

```python
#dropping unecessary columns
df = df.drop(['Application order', 'Nacionality', 'International', 'Inflation rate', 'GDP', 'Unemployment rate', 'Displaced'], axis=1)
df
```

```python
#creating dataframe with only graduate and dropout students
df = df[df.Target != 'Enrolled']
df
```

```python
#changing categroical values to numerical
# df['Target'] = pd.factorize(df['Target'])[0]
# df
```

```python
df.shape
```

```python
target = dict(
    Graduate=0,
    Dropoute=1,
)
```

```python
df
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df["target"] = df.Target.map(target)
print(df.Target.value_counts())
```

```python
model = smf.ols("df.iloc[:,24] ~ Target", data=df).fit()
print(model.summary())
```

```python
graduate_mean = df[df.Target=="Graduate"].iloc[:,24].mean()
dropout_mean = df[df.Target=="Dropout"].iloc[:,24].mean()
print(f"Average Graduate Curricular units 2nd sem (approved): {graduate_mean:0.0}")
print(f"Average Dropout Curricular units 2nd sem (approved): {dropout_mean:0.0}")
# print(f"Average difference: {dropout_mean - graduate_mean:0.0}")
```

```python
df = df.sort_values(by="Curricular units 2nd sem (approved)")
df["prediction"] = model.predict(df.Target)
```

```python
scatter = df.hvplot.scatter(x="Curricular units 2nd sem (approved)", y="prediction", by="Target")
scatter
```
